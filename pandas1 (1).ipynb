```python
df.dropna()
```

```python
df.year
```

```python
df.junetosepmm.sum()
```

```python
df.junetosepmm.fillna(df.junetosepmm.mean())
```

```python
df.junetosepmm.fillna(df.junetosepmm.mean(),inplace=True)
```

```python
df
```

```python
df.insert(2,column='junetosepcm',value=df.junetosepmm.div(10))
```

```python
df
```

```python
df.insert(3,column='prev_year_diff',value=df.junetosepmm.diff())
```

```python
df
```

```python
df.prev_year_diff.fillna(0,inplace=True)
```

```python
df
```

```python
df.insert(4,column='percentage_inc',value=df.prev_year_diff.pct_change().mul(100))
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
year=df.year
```

```python
rain=df.junetosepcm
```

```python
plt.plot(year,rain)
plt.figure(figsize=(10,3))
plt.show()
```

```python
df.plot(x='year',y='junetosepcm',figsize=(12,4),grid=True,xticks=year)
```

```python
plt.barh(year,rain)
plt.title('ANNUAL RAINFALL INDIA')
plt.xlabel('year')
plt.ylabel('rain in cms')
plt.show()
```

```python
df.plot(x='year',y='junetosepcm',figsize=(12,4),grid=True,kind='bar',
            xlabel='Year',ylabel='Rain in Cms',title='Rainall India June to Sep in Cms',
           color='pink')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1,figsize=(10,4))

change=df.prev_year_diff

axs[0].plot(year, rain)
axs[0].set_title('Rainfall in cms')

axs[1].bar(year, change)
axs[1].set_title('Change in cms')

plt.tight_layout()
plt.show()
```
